```python
import cv2

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    blur1 = cv2.GaussianBlur(frame, (7,7), 2)#больше размер - больше размытие, больше сигма - больше размытие (0-1)
    blur2 = cv2.GaussianBlur(frame, (21,21), 2)
    cv2.imshow('blur1',blur1)
    cv2.imshow('blur2', blur2)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
kernel1 = np.array([[-1,-2,-1],[-1,9,-1],[-1,-1,-1]])
kernel2 = np.ones((5, 5), np.float32)/30
while True:
    ret, frame = cap.read()
    dst1 = cv2.filter2D(frame, -1, kernel1)
    dst2 = cv2.filter2D(frame, -1, kernel2)
    cv2.imshow('dst1', dst1)
    cv2.imshow('dst2', dst2)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
# создаем окно для отображения результата и бегунки
cv2.namedWindow("setup")
cv2.createTrackbar("b1", "setup", 0, 255, nothing)
cv2.createTrackbar("g1", "setup", 0, 255, nothing)
cv2.createTrackbar("r1", "setup", 0, 255, nothing)
cv2.createTrackbar("b2", "setup", 255, 255, nothing)
cv2.createTrackbar("g2", "setup", 255, 255, nothing)
cv2.createTrackbar("r2", "setup", 255, 255, nothing)

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    r1 = cv2.getTrackbarPos('r1', 'setup')
    g1 = cv2.getTrackbarPos('g1', 'setup')
    b1 = cv2.getTrackbarPos('b1', 'setup')
    r2 = cv2.getTrackbarPos('r2', 'setup')
    g2 = cv2.getTrackbarPos('g2', 'setup')
    b2 = cv2.getTrackbarPos('b2', 'setup')
    min_p = (g1, b1, r1)
    max_p = (g2, b2, r2)
    img_g = cv2.inRange(frame, min_p, max_p)

    cv2.imshow('img', img_g)
    
    if cv2.waitKey(33) & 0xFF == ord('q'):
         break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    dst = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    L_limit=np.array([98,50,50])
    U_limit=np.array([139,255,255])
    b_mask=cv2.inRange(dst,L_limit,U_limit)
    blue=cv2.bitwise_and(frame,frame,mask=b_mask)
    L2_limit=np.array([0,0,0])
    U2_limit=np.array([0,150,150])
    bl_mask=cv2.inRange(dst,L2_limit,U2_limit)
    black=cv2.bitwise_and(frame,frame,mask=bl_mask)
    gen=cv2.bitwise_or(blue,black,mask=None)
    cv2.imshow('frame', frame)
    cv2.imshow('dst', gen)#blue
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    dst = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    L_limit=np.array([98,50,50])
    U_limit=np.array([139,255,255])
    b_mask=cv2.inRange(dst,L_limit,U_limit)
    moments = cv2.moments(b_mask, 1)
    dM01 = moments['m01']
    dM10 = moments['m10']
    dArea = moments['m00']
    if dArea > 150:
        x = int(dM10/dArea)
        y = int(dM01/dArea)
        cv2.circle(frame, (x,y), 10, (0,0,255), -1)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np 
  
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
    edged = cv2.Canny(gray, 30, 200) 
    contours, hierarchy = cv2.findContours(edged,  
        cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 
    cv2.imshow('Canny Edges After Contouring', edged) 
    cv2.drawContours(frame, contours, -1, (0, 255, 0), 3) 
    cv2.imshow('Contours', frame) 
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows() 
```

```python
import cv2

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    height, width = frame.shape[:2]
    edge = 10
    dst = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    L_limit=np.array([98,50,50])
    U_limit=np.array([139,255,255])
    b_mask=cv2.inRange(dst,L_limit,U_limit)
    moments = cv2.moments(b_mask, 1)
    dM01 = moments['m01']
    dM10 = moments['m10']
    dArea = moments['m00']
    if dArea > 150:
        x = int(dM10/dArea)
        y = int(dM01/dArea)
        cv2.circle(frame, (x,y), 10, (0,0,255), -1)
    if (x>(width/2+edge)) and x!=0:
        cv2.rectangle(frame, (0,0), (30,height), (0,255,0), -1)
    if (x<(width/2-edge)) and x!=0:
        cv2.rectangle(frame, (width-30,0), (width,height), (0,255,0), -1)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    height, width = frame.shape[:2]
    edge = 10
    dst = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    L_limit=np.array([98,50,50])
    U_limit=np.array([139,255,255])
    b_mask=cv2.inRange(dst,L_limit,U_limit)
    moments = cv2.moments(b_mask, 1)
    dM01 = moments['m01']
    dM10 = moments['m10']
    dArea = moments['m00']
    if dArea > 150:
        x = int(dM10/dArea)
        y = int(dM01/dArea)
        cv2.circle(frame, (x,y), 10, (0,0,255), -1)
        cv2.circle(frame, (x,y/2), 10, (0,0,255), -1)
    if (x>(width/2+edge)) and x!=0:
        cv2.rectangle(frame, (0,0), (x-50,height), (0,255,0), -1)
    if (x<(width/2-edge)) and x!=0:
        cv2.rectangle(frame, (x+50,0), (width,height), (0,255,0), -1)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
